```python
import sqlite3

# Conectarse a la base de datos (o crearla si no existe)
conn = sqlite3.connect('empleados.db')

# Crear un cursor
cursor = conn.cursor()

# Crear la tabla de empleados
cursor.execute('''
CREATE TABLE IF NOT EXISTS empleados (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT NOT NULL,
    sueldo REAL NOT NULL,
    fecha_ingreso TEXT NOT NULL,
    puesto TEXT NOT NULL,
    departamento TEXT NOT NULL
)
''')

# Confirmar los cambios
conn.commit()

# Cerrar la conexión
conn.close()

print("Base de datos y tabla de empleados creadas exitosamente.")
```

```python
system_gpt = """
Eres un experto administrador de bases de datos SQL Lite.

El usuario te hará una pregunta y tu generarás un SQL Querie que le permita obtener la respuesta en base al contenido de la base de datos empleados.db.

Esta es la estructura de la tabla empleados:
CREATE TABLE IF NOT EXISTS empleados (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT NOT NULL,
    sueldo REAL NOT NULL,
    fecha_ingreso TEXT NOT NULL,
    puesto TEXT NOT NULL,
    departamento TEXT NOT NULL
)

Las fechas se guardan en el formato 'YYYY-MM-DD'.
Los departamentos son: 'Desarrollo', 'Diseño', 'Sistemas', 'Administración', 'Gerencia'. (Considera mayúsculas y tildes).

No des más explicaciones en tu respuesta.
No utilices triples comillas para responder.
Responde únicamente con el SQL Query.

Ejemplos de preguntas que el usuario podría hacerte:
1. ¿Cuántos empleados hay en total?

Respuesta esperada: "SELECT COUNT(*) FROM empleados"
"""
```

```python
pregunta = """

Quien es el empleado que tiene el sueldo mas alto, dame toda su información

"""

from openai import OpenAI
client = OpenAI(api_key=api_key)

response = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {
      "role": "system",
      "content": [
        {
          "type": "text",
          "text": system_gpt
        }
      ]
    },
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": pregunta
        }
      ]
    }
  ],
  temperature=0.1,
  max_tokens=2560,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

respuesta = response.choices[0].message.content
print(respuesta)

import sqlite3

# Conectarse a la base de datos
conn = sqlite3.connect('empleados.db')

# Crear un cursor
cursor = conn.cursor()

# Consultar datos de la tabla de empleados

cursor.execute(respuesta)

# Obtener todos los registros
empleados = cursor.fetchall()

# Imprimir los registros
for empleado in empleados:
    print(empleado)

# Cerrar la conexión
conn.close()
```
